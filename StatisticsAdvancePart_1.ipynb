Q1 - What is a random variable in probability theory?

A random variable is a function that assigns numerical values to outcomes of a random experiment. It can be:

1.)Discrete: Takes countable values (e.g., number of heads in coin tosses).

2.)Continuous: Takes uncountable values within a range (e.g., height or time).

It is described by a probability distribution (PMF for discrete, PDF for continuous) and characterized by measures like expected value (mean) and variance (spread).

Q2 - What are the types of random variables?

Random variables are classified into two main types:

1) Discrete Random Variables:-

These take on a countable number of distinct values.
Example: The number of heads when flipping a coin three times (0, 1, 2, or 3).
Probability is assigned to individual values (e.g., using a probability mass function, PMF).

2) Continuous Random Variables:-

These take on an infinite number of possible values within a given range.
Example: The height of students in a class or the time taken to complete a task.
Probability is assigned to intervals, not individual points (e.g., using a probability density function, PDF).

Q3 - > What is the difference between discrete and continuous distributions?

In discrete distributions, probability is assigned to specific values.

In continuous distributions, probability is assigned to intervals, and the probability of any single value is zero (since there are infinitely many possible values).

Q4 - What are probability distribution functions (PDF)?

PDF (Probability Density Function): Used for continuous random variables. Describes the likelihood of a variable falling within a range. Probabilities are calculated as areas under the curve.
P(X=x)=0 for any single value.

PMF (Probability Mass Function): Used for discrete random variables. Gives the probability of the variable taking a specific value.
P(X=x) can be non-zero.

Both describe how probabilities are distributed, but PDF is for continuous data, and PMF is for discrete data.

Q13 - What is the Central Limit Theorem (CLT), and why is it critical in statistics?

The Central Limit Theorem (CLT) states that, regardless of the population's original distribution, the sampling distribution of the sample mean approaches a normal distribution as the sample size (n) increases, provided n is sufficiently large (typically n≥30).

Why is CLT Critical?
✅ Enables Normal Approximation: Even if data isn't normally distributed, we can still use normal-based statistical methods.
✅ Foundation for Inferential Statistics: Supports hypothesis testing and confidence intervals.
✅ Justifies Sampling Methods: Allows us to make predictions about populations from samples.
✅ Widely Used in Machine Learning & Data Science: Essential for statistical modeling and predictive analysis.

Q17 - What are point estimates and interval estimates in statistics?

Point Estimates vs. Interval Estimates in Statistics
1. Point Estimate 📍
A point estimate is a single value used to approximate a population parameter. It provides a best guess but does not show variability.

✅ Example:

Sample mean estimates the population mean.
Sample proportion estimates population proportion.
🚀 Limitation: A point estimate does not account for uncertainty in the estimate.

2. Interval Estimate 📏
An interval estimate provides a range of values within which the population parameter is likely to lie. It accounts for sampling variability and is often expressed with a confidence level (e.g., 95%).

✅ Example:

A 95% confidence interval (CI) for the mean might be (50, 60), meaning the true mean is likely between 50 and 60.

Q3 - Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

```python
def bernoulli_pdf(p, x):
    """
    Calculate the probability distribution function (PDF) of a Bernoulli distribution.

    Parameters:
    p (float): Probability of success (0 <= p <= 1)
    x (int): The outcome (0 or 1)

    Returns:
    float: The probability of outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be either 0 or 1")

    return p**x * (1 - p)**(1 - x)

# Example usage:
p = 0.7  # Probability of success (e.g., probability of heads)
x_values = [0, 1]  # Possible outcomes (0 or 1)

for x in x_values:
    print(f"P(X = {x}) = {bernoulli_pdf(p, x)}")
```

Q6 - Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the discrete uniform distribution
low = 1  # Minimum value
high = 6  # Maximum value (exclusive)

# Generate the values of the discrete uniform distribution
values = np.arange(low, high)

# Calculate the probability mass function (PMF)
pmf = np.ones_like(values, dtype=float) / len(values)

# Calculate the CDF by taking the cumulative sum of the PMF
cdf = np.cumsum(pmf)

# Plot the CDF
plt.step(values, cdf, where='post', color='skyblue', label='CDF', linewidth=2)
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution Function (CDF) of Discrete Uniform Distribution')
plt.xticks(values)
plt.yticks(np.linspace(0, 1, 11))
plt.grid(True)
plt.legend()
plt.show()
```

Q9 - Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate the Z-scores for the given dataset.

    Parameters:
    data (list or numpy array): The input dataset.

    Returns:
    numpy array: Z-scores of the dataset.
    """
    # Calculate mean and standard deviation of the dataset
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev
    return z_scores

# Example usage:
# Generate a sample dataset (e.g., normally distributed data)
np.random.seed(42)
data = np.random.normal(0, 1, 1000)  # Mean=0, Std Dev=1, 1000 samples

# Calculate the Z-scores for the dataset
z_scores = calculate_z_scores(data)

# Plot the Z-scores
plt.figure(figsize=(10, 6))
plt.plot(z_scores, linestyle='-', marker='o', color='skyblue', alpha=0.7, markersize=3)
plt.title('Z-scores of the Dataset')
plt.xlabel('Index')
plt.ylabel('Z-score')
plt.grid(True)
plt.show()
```

Q10 - Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the exponential distribution
lambda_ = 1  # Rate parameter (mean = 1 / lambda)
num_samples = 1000  # Number of samples to generate
sample_size = 30  # Size of each sample (to observe the CLT)

# Generate data from the exponential distribution
data = np.random.exponential(1/lambda_, num_samples)

# Function to apply the Central Limit Theorem by taking sample means
def apply_clt(data, sample_size, num_samples):
    sample_means = []
    for _ in range(num_samples):
        sample = np.random.choice(data, sample_size)
        sample_means.append(np.mean(sample))
    return np.array(sample_means)

# Apply CLT
sample_means = apply_clt(data, sample_size, num_samples)

# Plot the histogram of the original non-normal distribution (Exponential)
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='black', color='skyblue', alpha=0.7)
plt.title('Original Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the histogram of the sample means (After applying CLT)
plt.subplot(1, 2, 2)
plt.hist(sample_means, bins=30, edgecolor='black', color='skyblue', alpha=0.7)
plt.title('Sample Means (CLT Applied)')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Q18 - Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data, x):
    """
    Calculate the Z-score for a given data point x.

    Parameters:
    data (list or numpy array): The dataset
    x (float): The data point to calculate the Z-score for

    Returns:
    float: The Z-score
    """
    # Calculate mean and standard deviation of the dataset
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate the Z-score
    z_score = (x - mean) / std_dev
    return z_score

def compare_z_to_normal(z_score):
    """
    Compare the Z-score to the standard normal distribution.

    Parameters:
    z_score (float): The Z-score to compare

    Returns:
    float: The cumulative probability from the standard normal distribution
    """
    # Calculate the cumulative probability (CDF) for the Z-score
    cdf = norm.cdf(z_score)
    return cdf

# Example usage
# Generate a sample dataset (e.g., normally distributed data)
np.random.seed(42)
data = np.random.normal(50, 10, 1000)  # Mean=50, Std Dev=10, 1000 samples

# Define the data point for which we want to calculate the Z-score
x = 60  # Example data point

# Calculate the Z-score for the data point
z_score = calculate_z_score(data, x)

# Compare the Z-score to the standard normal distribution
cdf_value = compare_z_to_normal(z_score)

# Print the results
print(f"Z-score for x = {x}: {z_score:.2f}")
print(f"Cumulative probability (CDF) for Z-score: {cdf_value:.4f}")

# Plot the standard normal distribution and indicate the Z-score on the plot
x_values = np.linspace(-4, 4, 1000)
y_values = norm.pdf(x_values)

plt.plot(x_values, y_values, label="Standard Normal Distribution", color='skyblue')
plt.axvline(x=z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.title(f"Standard Normal Distribution with Z-score ({z_score:.2f})")
plt.xlabel('Z-score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```

Q19 - Implement hypothesis testing using Z-statistics for a sample dataset.

```python
import numpy as np
from scipy.stats import norm

def z_test(sample, population_mean, population_std, alpha=0.05):
    """
    Perform a Z-test for a sample dataset against a population mean.

    Parameters:
    sample (array): The sample data
    population_mean (float): The population mean under the null hypothesis
    population_std (float): The population standard deviation
    alpha (float): The significance level (default is 0.05)

    Returns:
    Z-statistic, p-value, and conclusion (reject or fail to reject H₀)
    """
    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Z-statistic calculation
    z_statistic = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))  # Two-tailed test

    # Conclusion based on p-value and significance level
    if p_value < alpha:
        conclusion = "Reject the null hypothesis (H₀)"
    else:
        conclusion = "Fail to reject the null hypothesis (H₀)"

    return z_statistic, p_value, conclusion

# Example usage
# Generate a sample dataset (e.g., normally distributed data)
np.random.seed(42)
sample_data = np.random.normal(50, 10, 100)  # Mean=50, Std Dev=10, 100 samples

# Population parameters
population_mean = 52  # Hypothesized population mean (H₀: µ = 52)
population_std = 10   # Population standard deviation

# Perform the Z-test
z_stat, p_value, conclusion = z_test(sample_data, population_mean, population_std)

# Print the results
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Conclusion: {conclusion}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculate the confidence interval for a sample dataset.

    Parameters:
    data (array): The dataset
    confidence_level (float): The confidence level (default is 0.95)

    Returns:
    tuple: lower bound, upper bound of the confidence interval
    """
    # Sample statistics
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)  # Sample standard deviation (ddof=1 for sample std)
    sample_size = len(data)

    # Z-score for 95% confidence level (for a two-tailed test, it is approximately 1.96)
    z_score = norm.ppf(1 - (1 - confidence_level) / 2)

    # Standard error of the mean
    standard_error = sample_std / np.sqrt(sample_size)

    # Margin of error
    margin_of_error = z_score * standard_error

    # Confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
# Generate a sample dataset (e.g., normally distributed data)
np.random.seed(42)
data = np.random.normal(50, 10, 100)  # Mean=50, Std Dev=10, 100 samples

# Calculate the confidence interval
lower_bound, upper_bound = calculate_confidence_interval(data, confidence_level=0.95)

# Print the confidence interval
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")

# Plotting the data and the confidence interval
plt.figure(figsize=(10, 6))
plt.hist(data, bins=20, edgecolor='black', color='skyblue', alpha=0.7)
plt.axvline(x=lower_bound, color='red', linestyle='--', label=f'Lower Bound: {lower_bound:.2f}')
plt.axvline(x=upper_bound, color='green', linestyle='--', label=f'Upper Bound: {upper_bound:.2f}')
plt.axvline(x=np.mean(data), color='orange', linestyle='-', label=f'Sample Mean: {np.mean(data):.2f}')
plt.title('Confidence Interval for the Sample Mean')
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)
plt.show()
```

Q22 - Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def plot_normal_pdf(mean, std_dev):
    """
    Calculate and visualize the Probability Density Function (PDF) of a normal distribution.

    Parameters:
    mean (float): The mean of the normal distribution
    std_dev (float): The standard deviation of the normal distribution
    """
    # Generate x values from mean - 4*std_dev to mean + 4*std_dev (to capture most of the data)
    x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)

    # Calculate the PDF of the normal distribution for the generated x values
    pdf = norm.pdf(x, loc=mean, scale=std_dev)

    # Plot the PDF
    plt.plot(x, pdf, label=f'Normal Distribution (mean={mean}, std={std_dev})', color='skyblue')
    plt.fill_between(x, pdf, color='skyblue', alpha=0.5)
    plt.title('Probability Density Function (PDF) of Normal Distribution')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage: Plotting PDF for a normal distribution with mean=0 and std_dev=1
plot_normal_pdf(mean=0, std_dev=1)
```

Q23 - Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

def plot_poisson_cdf(lam, x_max):
    """
    Plot the CDF of a Poisson distribution for a given lambda (mean) and maximum x value.

    Parameters:
    lam (float): The mean (lambda) of the Poisson distribution
    x_max (int): The maximum value of x to plot the CDF
    """
    # Generate x values (number of events)
    x = np.arange(0, x_max + 1)

    # Calculate the CDF for each x value
    cdf_values = poisson.cdf(x, lam)

    # Plot the CDF
    plt.step(x, cdf_values, where='post', label=f'Poisson CDF (λ={lam})', color='skyblue', linewidth=2)
    plt.fill_between(x, cdf_values, color='skyblue', alpha=0.5)
    plt.title(f'Cumulative Distribution Function (CDF) of Poisson Distribution')
    plt.xlabel('Number of Events (x)')
    plt.ylabel('Cumulative Probability')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage:
lambda_val = 5  # Poisson distribution mean (λ)
x_max = 15  # Maximum value of x for which we calculate the CDF

# Plot the Poisson CDF
plot_poisson_cdf(lambda_val, x_max)

# Example: Calculate the Poisson CDF at x = 5
cdf_at_5 = poisson.cdf(5, lambda_val)
print(f"The CDF value at x = 5 for Poisson distribution with λ = {lambda_val} is: {cdf_at_5:.4f}")
```

Q24 - Simulate a random variable using a continuous uniform distribution and calculate its expected value.

```python
import numpy as np

def simulate_uniform_random_variable(a, b, num_samples):
    """
    Simulate a random variable using a continuous uniform distribution and calculate its expected value.

    Parameters:
    a (float): The lower bound of the distribution
    b (float): The upper bound of the distribution
    num_samples (int): The number of samples to simulate

    Returns:
    float: The expected value of the uniform distribution
    numpy array: The simulated random variables
    """
    # Simulate random variables from a continuous uniform distribution
    random_variables = np.random.uniform(a, b, num_samples)

    # Calculate the expected value (mean) of the uniform distribution
    expected_value = (a + b) / 2

    return expected_value, random_variables

# Example usage:
a = 0  # Lower bound of the uniform distribution
b = 10  # Upper bound of the uniform distribution
num_samples = 1000  # Number of random samples to simulate

# Simulate the random variable and calculate the expected value
expected_value, random_variables = simulate_uniform_random_variable(a, b, num_samples)

# Print the expected value
print(f"Expected Value (E(X)) for uniform distribution over [{a}, {b}]: {expected_value}")

# Print a few simulated random variables for illustration
print(f"Some simulated random variables: {random_variables[:10]}")
```

Q25 - Write a Python program to compare the standard deviations of two datasets and visualize the difference.

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_standard_deviations(data1, data2):
    """
    Compare the standard deviations of two datasets and visualize the difference.

    Parameters:
    data1 (array-like): The first dataset
    data2 (array-like): The second dataset

    Returns:
    tuple: standard deviation of data1, standard deviation of data2
    """
    # Calculate standard deviations
    std_dev1 = np.std(data1, ddof=1)  # Sample standard deviation (ddof=1)
    std_dev2 = np.std(data2, ddof=1)  # Sample standard deviation (ddof=1)

    return std_dev1, std_dev2

# Example usage:
# Generate two sample datasets
np.random.seed(42)

# Dataset 1: Normal distribution with mean=50, std=5
data1 = np.random.normal(50, 5, 1000)

# Dataset 2: Normal distribution with mean=50, std=10
data2 = np.random.normal(50, 10, 1000)

# Compare standard deviations
std_dev1, std_dev2 = compare_standard_deviations(data1, data2)

# Print the standard deviations
print(f"Standard Deviation of Dataset 1: {std_dev1:.2f}")
print(f"Standard Deviation of Dataset 2: {std_dev2:.2f}")

# Plotting the histograms of both datasets to visualize the difference
plt.figure(figsize=(12, 6))

# Histogram for Dataset 1
plt.subplot(1, 2, 1)
plt.hist(data1, bins=30, edgecolor='black', color='skyblue', alpha=0.7)
plt.title(f'Dataset 1: Std Dev = {std_dev1:.2f}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Histogram for Dataset 2
plt.subplot(1, 2, 2)
plt.hist(data2, bins=30, edgecolor='black', color='salmon', alpha=0.7)
plt.title(f'Dataset 2: Std Dev = {std_dev2:.2f}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()
```

Q26 - Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

```python
import numpy as np

def calculate_range_and_iqr(data):
    """
    Calculate the range and interquartile range (IQR) of a dataset.

    Parameters:
    data (array): The dataset

    Returns:
    tuple: range, IQR of the dataset
    """
    # Calculate the range (difference between max and min values)
    data_range = np.max(data) - np.min(data)

    # Calculate the IQR (difference between Q3 and Q1)
    Q1 = np.percentile(data, 25)  # 25th percentile (Q1)
    Q3 = np.percentile(data, 75)  # 75th percentile (Q3)
    iqr = Q3 - Q1

    return data_range, iqr

# Example usage:
# Generate a dataset from a normal distribution (mean=50, std=10, 1000 samples)
np.random.seed(42)
data = np.random.normal(50, 10, 1000)

# Calculate range and IQR
data_range, iqr = calculate_range_and_iqr(data)

# Print the results
print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr:.2f}")
```

Q27 - 5 Implement Z-score normalization on a dataset and visualize its transformation.

```python
import numpy as np
import matplotlib.pyplot as plt

def z_score_normalization(data):
    """
    Apply Z-score normalization to a dataset.

    Parameters:
    data (array): The dataset

    Returns:
    array: Z-score normalized data
    """
    # Calculate mean and standard deviation of the dataset
    mean = np.mean(data)
    std_dev = np.std(data)

    # Z-score normalization
    normalized_data = (data - mean) / std_dev

    return normalized_data

# Example usage:
# Generate a dataset (e.g., normally distributed data)
np.random.seed(42)
data = np.random.normal(50, 10, 1000)  # Mean=50, Std Dev=10, 1000 samples

# Apply Z-score normalization
normalized_data = z_score_normalization(data)

# Plotting the original and normalized data
plt.figure(figsize=(14, 6))

# Original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, edgecolor='black', color='skyblue', alpha=0.7)
plt.title('Original Dataset')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)

# Normalized data
plt.subplot(1, 2, 2)
plt.hist(normalized_data, bins=30, edgecolor='black', color='salmon', alpha=0.7)
plt.title('Z-Score Normalized Dataset')
plt.xlabel('Z-score')
plt.ylabel('Frequency')
plt.grid(True)

# Show the plots
plt.tight_layout()
plt.show()

```

Q28 - Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal
distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_and_kurtosis(data):
    """
    Calculate skewness and kurtosis of a dataset.

    Parameters:
    data (array): The dataset

    Returns:
    tuple: skewness, kurtosis of the dataset
    """
    # Calculate skewness
    skew_value = skew(data)

    # Calculate kurtosis (excess kurtosis, i.e., normal distribution kurtosis is 0)
    kurtosis_value = kurtosis(data)

    return skew_value, kurtosis_value

# Example usage:
# Generate a dataset from a normal distribution (mean=0, std=1, 1000 samples)
np.random.seed(42)
data = np.random.normal(0, 1, 1000)  # Mean=0, Std Dev=1, 1000 samples

# Calculate skewness and kurtosis
skew_value, kurtosis_value = calculate_skewness_and_kurtosis(data)

# Print the results
print(f"Skewness of the dataset: {skew_value:.4f}")
print(f"Kurtosis of the dataset: {kurtosis_value:.4f}")
```
